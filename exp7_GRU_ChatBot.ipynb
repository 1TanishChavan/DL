```python

conversations = [
    ("Hi", "Hello! How can I help you?"),
    ("What is your name?", "Rugved"),
    ("What is your name?", "Harsh"),
    ("What is your name?", "Mayuresh"),
    ("How are you?", "I'm an AI, I don't have feelings, but thanks for asking!"),
    ("Tell me a joke", "Why donâ€™t scientists trust atoms? Because they make up everything!"),
    ("Goodbye", "Goodbye! Have a great day!"),
    ("What's the weather like?", "It's sunny and warm today."),
    ("Can you help me with my homework?", "Sure, what subject are you working on?"),
    ("What's your favorite color?", "I don't have preferences, but I think blue is nice."),
    ("Do you like music?", "I don't have feelings, but I can help you find some music!"),
    ("What's the capital of France?", "The capital of France is Paris."),
    ("How old are you?", "I don't have an age, I'm just a program."),
    ("What's your favorite food?", "I don't eat, but I hear pizza is popular!"),
    ("Can you tell me a fun fact?", "Did you know that honey never spoils? Archaeologists have found pots of honey in ancient Egyptian tombs that are over 3,000 years old and still edible."),
    ("What's the time?", "It's 10:00 AM."),
    ("Do you have any hobbies?", "I don't have hobbies, but I can help you find one!"),
    ("What's your favorite movie?", "I don't watch movies, but I can recommend some!"),
    ("Can you speak other languages?", "Yes, I can understand and respond in multiple languages."),
    ("What's your purpose?", "I'm here to assist you with information and tasks."),
    ("What's your favorite book?", "I don't read, but I can suggest some popular titles."),
    ("Can you play games?", "I can help you find games to play!"),
    ("What's your favorite animal?", "I don't have preferences, but many people love dogs."),
    ("Do you have a family?", "I don't have a family, I'm just a program."),
    ("What's your favorite sport?", "I don't play sports, but I can give you information about them."),
    ("Can you tell me a story?", "Once upon a time, in a land far, far away..."),
    ("What's your favorite song?", "I don't listen to music, but I can suggest some popular songs."),
    ("Do you know any poems?", "Sure, here's a short one: Roses are red, violets are blue..."),
    ("What's your favorite season?", "I don't experience seasons, but many people love spring."),
    ("Can you solve math problems?", "Yes, I can help with math problems. What do you need help with?"),
    ("What's your favorite holiday?", "I don't celebrate holidays, but many people love Christmas."),
    ("Do you have any pets?", "I don't have pets, but I can give you information about them."),
    ("What's your favorite TV show?", "I don't watch TV, but I can suggest some popular shows."),
    ("Can you dance?", "I can't dance, but I can help you find some dance tutorials!"),
    ("What's your favorite game?", "I don't play games, but I can suggest some popular ones."),
    ("Do you like to travel?", "I don't travel, but I can give you information about different places."),
    ("What's your favorite quote?", "I don't have a favorite quote, but here's one: 'The only limit to our realization of tomorrow is our doubts of today.' - Franklin D. Roosevelt"),
    ("Can you cook?", "I can't cook, but I can help you find recipes!"),
    ("What's your favorite drink?", "I don't drink, but many people love coffee."),
    ("Do you like art?", "I don't have preferences, but I can help you find information about art."),
    ("What's your favorite subject?", "I don't have a favorite subject, but I can help you with many topics."),
    ("Can you sing?", "I can't sing, but I can help you find some music!"),
    ("What's your favorite place?", "I don't have a favorite place, but many people love the beach."),
    ("Do you like reading?", "I don't read, but I can suggest some popular books."),
    ("What's your favorite hobby?", "I don't have hobbies, but I can help you find one!"),
    ("Can you write a poem?", "Sure, here's a short one: The sun sets in the west, bringing the day to rest."),
    ("What's your favorite fruit?", "I don't eat, but many people love apples."),
    ("Do you like puzzles?", "I don't solve puzzles, but I can help you find some!"),
    ("What's your favorite flower?", "I don't have preferences, but many people love roses."),
    ("Can you tell me a riddle?", "Sure, here's one: What has keys but can't open locks? A piano."),
    ("What's your favorite vegetable?", "I don't eat, but many people love carrots."),
    ("Do you like science?", "I don't have preferences, but I can help you with science topics."),
    ("What's your favorite planet?", "I don't have preferences, but many people love Earth."),
    ("Can you tell me a proverb?", "Sure, here's one: 'A stitch in time saves nine.'"),
    ("What's your favorite constellation?", "I don't have preferences, but many people love Orion."),
    ("Do you like history?", "I don't have preferences, but I can help you with history topics."),
    ("What's your favorite invention?", "I don't have preferences, but many people love the internet."),
    ("Can you tell me a myth?", "Sure, here's one: The myth of Icarus, who flew too close to the sun."),
    ("What's your favorite constellation?", "I don't have preferences, but many people love Orion."),
    ("Do you like geography?", "I don't have preferences, but I can help you with geography topics."),
    ("What's your favorite landmark?", "I don't have preferences, but many people love the Eiffel Tower."),
    ("Can you tell me a legend?", "Sure, here's one: The legend of King Arthur and the Knights of the Round Table."),
    ("What's your favorite constellation?", "I don't have preferences, but many people love Orion."),
    ("Do you like astronomy?", "I don't have preferences, but I can help you with astronomy topics."),
    ("What's your favorite star?", "I don't have preferences, but many people love the Sun."),
    ("Can you tell me a fable?", "Sure, here's one: The Tortoise and the Hare."),
    ("What's your favorite constellation?", "I don't have preferences, but many people love Orion."),
    ("Do you like mythology?", "I don't have preferences, but I can help you with mythology topics."),
    ("What's your favorite mythological creature?", "I don't have preferences, but many people love dragons."),
    ("Can you tell me a fairy tale?", "Sure, here's one: Cinderella."),
    ("What's your favorite constellation?", "I don't have preferences, but many people love Orion."),
    ("Do you like folklore?", "I don't have preferences, but I can help you with folklore topics."),
    ("What's your favorite folk tale?", "I don't have preferences, but many people love the story of Robin Hood."),
    ("Can you tell me a ghost story?", "Sure, here's one: The story of the haunted house on the hill."),
    ("What's your favorite constellation?", "I don't have preferences, but many people love Orion."),
    ("Do you like legends?", "I don't have preferences, but I can help you with legends."),
    ("What's your favorite legendary hero?", "I don't have preferences, but many people love Hercules."),
    ("Can you tell me a tall tale?", "Sure, here's one: The story of Paul Bunyan."),
    ("What's your favorite constellation?", "I don't have preferences, but many people love Orion."),
    ("Do you like epic stories?", "I don't have preferences, but I can help you with epic stories."),
    ("What's your favorite epic?", "I don't have preferences, but many people love the Odyssey."),
    ("Can you tell me a saga?", "Sure, here's one: The saga of Beowulf."),
    ("What's your favorite constellation?", "I don't have preferences, but many people love Orion."),
    ("Do you like heroic tales?", "I don't have preferences, but I can help you with heroic tales."),
    ("What's your favorite hero?", "I don't have preferences, but many people love Superman."),
    ("Can you tell me a heroic story?", "Sure, here's one: The story of Perseus and Medusa."),
    ("What's your favorite constellation?", "I don't have preferences, but many people love Orion."),
    ("Do you like adventure stories?", "I don't have preferences, but I can help you with adventure stories."),
    ("What's your favorite adventure?", "I don't have preferences, but many people love the adventures of Sherlock Holmes."),
    ("Can you tell me an adventure story?", "Sure, here's one: The story of Indiana Jones."),
    ("What's your favorite constellation?", "I don't have preferences, but many people love Orion."),
    ("Do you like mystery stories?", "I don't have preferences, but I can help you with mystery stories."),
    ("What's your favorite mystery?", "I don't have preferences, but many people love the mystery of the Bermuda Triangle."),
    ("Can you tell me a mystery story?", "Sure, here's one: The story of the missing treasure."),
    ("What's your favorite constellation?", "I don't have preferences, but many people love Orion.")
]
import pandas as pd
extended_conversations = pd.DataFrame(conversations, columns=['Input', 'Response'])
extended_conversations.to_csv('extended_conversations.csv', index=False)
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, GRU, Dense, Dropout

# Load the dataset
data = pd.read_csv('/content/extended_conversations.csv')

# Prepare the input and output sequences
input_texts = data['Input'].values
target_texts = data['Response'].values

# Tokenize the input and output texts
tokenizer = Tokenizer()
tokenizer.fit_on_texts(input_texts + target_texts)
input_sequences = tokenizer.texts_to_sequences(input_texts)
target_sequences = tokenizer.texts_to_sequences(target_texts)

# Pad the sequences
max_sequence_length = max([len(seq) for seq in input_sequences + target_sequences])
input_sequences = pad_sequences(input_sequences, maxlen=max_sequence_length, padding='post')
target_sequences = pad_sequences(target_sequences, maxlen=max_sequence_length, padding='post')

# Create the target data for training
target_sequences = np.array(target_sequences)
target_sequences = np.expand_dims(target_sequences, -1)
# Define model parameters
vocab_size = len(tokenizer.word_index) + 1
embedding_dim = 100
gru_units = 256

# Build the model
model = Sequential()
model.add(Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=max_sequence_length))
model.add(GRU(units=gru_units, return_sequences=True))
model.add(Dropout(0.2))
model.add(Dense(units=vocab_size, activation='softmax'))

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.summary()
# Train the model
batch_size = 64
epochs = 200

model.fit(input_sequences, target_sequences, batch_size=batch_size, epochs=epochs, validation_split=0.2)
def generate_response(input_text):
    input_sequence = tokenizer.texts_to_sequences([input_text])
    input_sequence = pad_sequences(input_sequence, maxlen=max_sequence_length, padding='post')
    predicted_sequence = model.predict(input_sequence)
    predicted_text = tokenizer.sequences_to_texts(predicted_sequence.argmax(axis=-1))
    return predicted_text[0]

# Example usage
response = generate_response("Tell me a joke")
print(response)
```
